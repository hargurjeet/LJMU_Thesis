```python
import torch
from transformers import pipeline

def initialize_pipeline(model_id):
    """Initialize the text generation pipeline."""
    text_generation_pipeline = pipeline(
        "text-generation",
        model=model_id,
        model_kwargs={"torch_dtype": torch.bfloat16},
        device_map="auto",
    )
    return text_generation_pipeline

# Model ID
model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

# Initialize the pipeline
text_generation_pipeline = initialize_pipeline(model_id)
```

```python
def generate_text(pipeline, messages, max_new_tokens=256):
    """Generate text based on the input messages."""
    outputs = pipeline(
        messages,
        max_new_tokens=max_new_tokens,
    )
    return outputs[0]["generated_text"][-1]['content']

messages1 = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

generated_text1 = generate_text(text_generation_pipeline, messages1)
print(f"Generated text for first prompt: {generated_text1}")
```

```python
generated_texts = []
for _, row in customer_data.iterrows():
    messages = [
        {"role": "system", "content": "You are an AI assistant with good knowledge of the travel industry.\
        A Tour & Travels Company Wants To Predict Whether A Customer Will Churn Or Not Based On Indicators Given Below.\
        Convert the following row to a detailed textual description:\
        Consider the following constraints: 1.Dont repeat similar discription for every line. No need to follow an order\
        while generation of discription which mean the age can go in last or annual income can be in first...etc.\
        dont include this (Here's a detailed textual description of the given row:) for each response.\
        "},
        {"role": "user", "content": f"Age: {row['Age']}, FrequentFlyer: {row['FrequentFlyer']}, AnnualIncomeClass: {row['AnnualIncomeClass']}, ServicesOpted: {row['ServicesOpted']}, \
        AccountSyncedToSocialMedia: {row['AccountSyncedToSocialMedia']}, BookedHotelOrNot: {row['BookedHotelOrNot']}"}
    ]
    generated_text = generate_text(text_generation_pipeline, messages)
    generated_texts.append(generated_text)

customer_data['GeneratedText'] = generated_texts
```

```python
# Function to generate text for a batch of messages
def batch_generate_text(pipeline, messages, max_new_tokens=50):
    """Generate text for a batch of messages."""
    # Prepare the prompts for batch generation
    prompts = []
    for message in messages:
        prompt = (
            "You are an AI assistant with good knowledge of the travel industry. "
            "A Tour & Travels Company Wants To Predict Whether A Customer Will Churn Or Not Based On Indicators Given Below. "
            "Convert the following row to a detailed textual description: "
            "Consider the following constraints: 1. Don't repeat similar descriptions for every line. "
            "No need to follow an order while generating the description which means the age can go in last or annual income can be in first...etc. "
            "Don't include this (Here's a detailed textual description of the given row:) for each response.\n"
            f"Age: {message['Age']}, FrequentFlyer: {message['FrequentFlyer']}, AnnualIncomeClass: {message['AnnualIncomeClass']}, "
            f"ServicesOpted: {message['ServicesOpted']}, AccountSyncedToSocialMedia: {message['AccountSyncedToSocialMedia']}, "
            f"BookedHotelOrNot: {message['BookedHotelOrNot']}"
        )
        prompts.append(prompt)

    # Generate text in batches using max_new_tokens
    generated_texts = pipeline(prompts, max_new_tokens=max_new_tokens)
    return [text['generated_text'].strip() for text in generated_texts]
```

```python
# Prepare messages for batch processing
messages = customer_data.to_dict(orient='records')

# Generate texts in a batch
generated_texts = batch_generate_text(text_generation_pipeline, messages)

# Adding generated texts back to DataFrame
customer_data['GeneratedText'] = generated_texts
```
