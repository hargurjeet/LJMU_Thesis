```python
import torch
from transformers import pipeline

def initialize_pipeline(model_id):
    """Initialize the text generation pipeline."""
    text_generation_pipeline = pipeline(
        "text-generation",
        model=model_id,
        model_kwargs={"torch_dtype": torch.bfloat16},
        device_map="auto",
    )
    return text_generation_pipeline

# Model ID
model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

# Initialize the pipeline
text_generation_pipeline = initialize_pipeline(model_id)
```

```python
def generate_text(pipeline, messages, max_new_tokens=256):
    """Generate text based on the input messages."""
    outputs = pipeline(
        messages,
        max_new_tokens=max_new_tokens,
    )
    return outputs[0]["generated_text"][-1]['content']

messages1 = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

generated_text1 = generate_text(text_generation_pipeline, messages1)
print(f"Generated text for first prompt: {generated_text1}")
```
